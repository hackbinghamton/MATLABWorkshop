# How to Access MATLAB (<10 minute total download and set-up time)

Everyone at Binghamton University should have access to MATLAB through our VMware Horizon Client. Many classes use this, and if you've taken one of these classes before, you should already have access to MATLAB through your account. If you have never used the VMware Horizon Client, you can follow the steps below!

---

## Setting up the VPN

Depending on where you are, you may need to use a VPN in order to access the VMware Horizon Client. Fortunately, Binghamton University has an easy-to-set-up VPN with a convenient site with pictures and instructions for both Mac and Windows!

[Pulse VPN](https://binghamton.service-now.com/sp?id=kb_article&sysparm_article=KB0010459&sys_kb_id=c9cb558cdb6f8410de8df4641f9619a8&spa=1)

---

## Setting Up VMware Horizon Client

From this point on, you can follow this PDF from Binghamton University: [Accessing MATLAB](https://www.binghamton.edu/its/about/organization/operations-and-infrastructure/networking/img/step3-access-using-bingview1020.pdf)

---

**If for whatever reason, you are not able to access the PDF, we have instructions here:**

Click the following link: [VMware Horizon Client](https://bingview.binghamton.edu/)

This should take you to a site where you can download the VMware Horizon Client best suited to your machine. (It can be downloaded for Windows, Mac, Linux, Android, iOS, and Chrome, although it is not recommended for Android or iOS.)



1.PNG

Pick the one most fitted for your computer. When it has finished installing, accept the default settings.

You may be prompted to restart your system depending on your device settings.

Start the VMware Horizon client and select New Server.

Enter the name: `bingview.binghamton.edu` and select `Connect`

2.PNG

Log in using your PODS username and password.

3.PNG

You should see a folder titled "MATLAB" on the desktop. (If not, you can search for it.)

4.PNG

Hopefully you were able to access MATLAB! For our next section, Navigating the MATLAB Environment, click here!
